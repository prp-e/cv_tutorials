```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../ep1/Lenna.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
blue_image = image[:,:,2]
```

```python
plt.imshow(blue_image)
```

```python
cropped_image = image[128:255, 128:255]
```

```python
plt.imshow(cropped_image)
```

```python
cropped_image.shape
```

```python
size = (1024, 1024)
resized_image = cv2.resize(image, size)
```

```python
plt.imshow(resized_image)
```

```python
resized_image.shape
```

```python
new_image = np.zeros([1024,1024,3])
```

```python
plt.imshow(new_image)
```

```python
new_image[:,:,0] = resized_image[:, :, 0]
```

```python
plt.imshow(new_image)
```

```python
new_image = np.zeros([128, 128, 3])
```

```python
plt.imshow(new_image)
```

```python
resized_image[0:128, 0:128] = new_image
```

```python
plt.imshow(resized_image)
```
